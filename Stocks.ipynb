```python
import pandas as pd
from requests import get
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
API_KEY=os.environ.get('BRAPIKEY')
```

```python
class Stock:
    def __init__(self, ticker):
        self.ticker = ticker
        self.url = f'https://brapi.dev/api/quote/{self.ticker}'
    
    def GetStockInfo(self):
        params = {
            'modules': 'summaryProfile',
            'token': API_KEY,}
        url = self.url
        try:
            req = get(url, params=params)
            data = req.json()
            df = pd.DataFrame(data['results'])
            info = pd.json_normalize(df['summaryProfile'])
            return info
        except Exception as e:
            print(e)


    def GetStockQuote(self):
        params = {
            'token': API_KEY}
        try:
            req = get(self.url, params=params)
            data = req.json()
            return pd.DataFrame(data['results'])
        except Exception as e:
            print(e)      

    def GetHistoricalQuote(self):
        params = {
            'token': API_KEY,
            'range':'3mo',
            'interval':'1d',}
        try:
            req = get(self.url, params=params)
            data = req.json()
            results = pd.json_normalize(data,  'results')
            history = results['historicalDataPrice'][0]
            df = pd.DataFrame(history)
            df['date'] = df['date'].astype('datetime64[s]')
            df['ticker'] = self.ticker
            return df
        except Exception as e:
            print(e)


    def TransformQuotes(data):   
        df = pd.DataFrame(data['results'])
        return df

```
